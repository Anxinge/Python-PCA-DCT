# Implement PCA for Dimensions Reduction 

```python
# load PCA libraries 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
# Standardize the feature matrix
X_1 = StandardScaler().fit_transform(df1)
# Create a PCA that will retain 90% of the variance
pca = PCA(n_components=0.9, whiten=True)

# Conduct PCA
X_pca_1= pca.fit_transform(X_1)
```

```python
# Show results
print('Original number of features for dataset1 :', X_1.shape[1])
print('Reduced number of features for dataset1:', X_pca_1.shape[1])
print("\n ** As the result of PCA, we can reduce the number of principal components to 71 from 100. ** \n")
```

```python
print(X_1.shape)
print(X_pca_1.shape)
print("\n ** So second dataset's shape is to change from (1000 * 100) to (1000 * 69) ** \n")
X_pca_1
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
# Standardize the feature matrix
X_2 = StandardScaler().fit_transform(df2)
# Create a PCA that will retain 90% of the variance
pca = PCA(n_components=0.9, whiten=True)

# Conduct PCA
X_pca_2= pca.fit_transform(X_2)
```

```python
# Show results
print('Original number of features for dataset1 :', X_2.shape[1])
print('Reduced number of features for dataset1:', X_pca_2.shape[1])
print("\n ** As the result of PCA, we can reduce the number of principal components to 70 from 100. ** \n")
```

```python
print(X_2.shape)
print(X_pca_2.shape)
X_pca_2
print("\n ** So second dataset's shape is to change from (1000 * 100) to (1000 * 69) ** \n")
X_pca_2
```

# Calculate the suitable number of components for PCA. 

```python
# determine the number of principal components for PCA.
pca = PCA().fit(df1)
print("\n** Cumulative explained variance via Numbers of components **\n")
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.show()
print("\n ** From this plot, we can know that the suitable number of principal components of the first data is 60 ~ 70.** \n")
```

# Use DCT method for features extraction.

```python
# Load tensorflow library to use DCT mothod.
import tensorflow as tf
print("\n ** As we knew the principal number of components is 70, so I have set the n = 70 as the DCT parameter **\n")
tf.compat.v1.signal.dct(
    df1.values,
    n = 70,
    type=1,
    axis=-1,
)
```

```python
import tensorflow as tf
tf.compat.v1.signal.dct(
    df2.values,
    n = 70,
    type=1,
    axis=-1,
)
```

 ** From above results, we can get the principal data (shape : (1000 *70)) using DCT method. The difference between two methods is to use Standardscalar technology or not. In PCA, we used the Standardscalar library to scalar dataset. it is needed to get result more accuracy. but in DCT method, there is no parameter to use scale dataset. so we can know that the result dataset is difference for each method.
 
To reduce the number of variables (ie: dimensions reduction),use PCA and DCT method for both dataset.
    as the result,for retain 90 % variance of the data, we can reduce the dimensions 100 to 70 using PCA.
    Also for the same result,we can set the DCT method's parameter n to 70.

By using all this method's of data mining and analyzing,we can reduce the amount of data to improve speed and memory of data processing.**
